```python
from collections import deque
```

```python
queue = deque()
```

```python
type(queue)
```

```python
# Enqueue
queue.append("A")
queue.append("B")
queue.append("C")
print(queue)   # deque(['A', 'B', 'C'])
```

```python
# Dequeue
front = queue.popleft()
```

```python
print(front)   # 'A'
print(queue)   # deque(['B', 'C'])
```
