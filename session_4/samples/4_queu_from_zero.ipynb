```python
class ColaManual:
    def __init__(self):
        # La lista de Python actúa como el contenedor de la cola
        self.items = []

    def is_empty(self):
        # Verifica si la cola está vacía
        return not self.items

    def enqueue(self, item):
        # Agregar al final (O(1))
        self.items.append(item)
        print(f"ENQUEUE: '{item}' agregado. Cola: {self.items}")

    def dequeue(self):
        # Eliminar del frente (¡Lento! O(n))
        if not self.is_empty():
            # El método pop(0) elimina el elemento en el índice 0 (el frente)
            item = self.items.pop(0)
            print(f"DEQUEUE: '{item}' atendido. Cola: {self.items}")
            return item
        else:
            print("Error: La cola está vacía.")
            return None
```

```python
print("\n--- Cola Manual ---")
cola = ColaManual()
```

```python
type(cola)
```

```python
cola.enqueue("Tarea 1")
cola.enqueue("Tarea 2")
cola.enqueue("Tarea 3")
```

```python
print("---")
cola.dequeue() # Saca Tarea 1
```

```python
cola.dequeue() # Saca Tarea 2
```
