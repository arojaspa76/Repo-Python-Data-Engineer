```python
class DynamicArray:  
    def __init__(self):
        self.capacity = 1  # capacidad inicial
        self.length = 0    # elementos actuales
        self.data = [None] * self.capacity

    def append(self, value): # 0 1 2
        # si est√° lleno, realocar con el doble de capacidad
        if self.length == self.capacity:
            self._resize(2 * self.capacity)
        self.data[self.length] = value
        self.length += 1

    def _resize(self, new_cap):
        new_data = [None] * new_cap
        for i in range(self.length):
            new_data[i] = self.data[i]
        self.data = new_data
        self.capacity = new_cap
```

```python
# Create the Dynamic Array Object
arr = DynamicArray()
```

```python
type(arr)
```

```python
for i in range(5):
    arr.append(i)
```

```python
print(f"Data: {arr.data}")
```

```python
print(f"Length: {arr.length}")
```

```python
print(f"Capacity: {arr.capacity}")
```
